```python
from diffusers import StableDiffusionPipeline
import torch

# 检查是否有可用的 GPU，如果没有则使用 CPU
device = "cuda" if torch.cuda.is_available() else "cpu"

# 加载模型（首次运行时会自动下载模型）
model_id = "CompVis/stable-diffusion-v1-4"
pipeline = StableDiffusionPipeline.from_pretrained(model_id, use_auth_token=True).to(device)

# 定义您的文本提示
prompt = "A fantasy landscape, vivid colors and detailed"

# 生成图像
with torch.no_grad():
    image = pipeline(prompt, num_inference_steps=50)["sample"][0]  # num_inference_steps 越高，生成的图像质量越好

# 保存或显示图像
image.save("generated_image.png")
```
